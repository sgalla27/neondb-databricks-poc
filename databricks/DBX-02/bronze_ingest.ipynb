```python
#read in policies.csv
df_policy = spark.read.option('header','true').csv('insurance_db/Volumes/insurance_files/policies.csv')
```

```python
#put policies.csv into bronze table
df_policy.write.format('delta').saveAsTable('insurance_db.bronze.policies')
```

```python
#read in claims.csv
df_policy = spark.read.option('header','true').csv('insurance_db/Volumes/insurance_files/claims.csv')
```

```python
#put claims.csv into bronze table
df_policy.write.format('delta').saveAsTable('insurance_db.bronze.claims')
```

```python
#show policies bronze table
spark.sql('SELECT * FROM insurance_db.bronze.policies').show()
```

```python
#show claims bronze table
spark.sql('SELECT * FROM insurance_db.bronze.claims').show()
```
