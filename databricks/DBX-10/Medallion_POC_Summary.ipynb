## Section 1: What is Medallion Architecture
Medallion Architecture is the data engineering design pattern where the data gets organized during layers which allows the data to get incrementally more clean.  

 It starts in the bronze layer where the raw data gets processed into tables without any cleaning. Then, it moves to the silver layer where the data gets cleaned by removing duplicates, casting types, and validate and standardize the data. Finally, with the gold layer it aggregates data and allows joins between tables.
<br/>

## Section 6: Free Edition limitations encountered
The limitations are there's no Delta Live Tables which automates Data pipelines. Also, there is no scala which is the programing language Spark code is in. Additionally, there's no processing data continuously as it arrives. Next, there is no cloud mounts which connects cloud storage into the workspace. Finally, there is limited compute which means that there is not a lot of power to run processes concurrently and queries need to be optimized.
<br/>
<br/>
## Section 7: How would this work in production?
Delta Live Tables would be used to automate pipeline orchestration, dependency management, and dataquality rules. Overall it simplifies monitoring and maintenance of Bronze-to-Silver-to-Gold pipelines.
<br/>
<br/>
Streaming for Near Real-Time Data allows for data to be processed as it is fed in instead of having to be fed in through batch loads. This allows analytics and dashboards to stay current.
<br/>
<br/>
Partitioning is important for performance because it allows for data to be split into subsections to be searched rather than searching the whole data, especially useful when there are millions and billions of records. Smae thing with incremental processing where if there are so many records that reprocessing them all day is expensive and slow, only process a certain amount of new/changed records.
<br/>
<br/>
Data Quality Controls allows for any bad data records to be quarantined for review rather than inserting it and having to backtrack and review and possibly remove later.
<br/>
<br/>
Monitoring and Alerting allows issues to be detected before any users notice and affects production, making it so issues are less fatal.
